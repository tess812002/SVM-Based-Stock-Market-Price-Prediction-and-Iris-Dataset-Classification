```python
import seaborn as sns
df=sns.load_dataset('iris')
df
```

```python
df["species"].value_counts()
```

```python
df.species=df.species.map({'setosa':1,"versicolor":2,"virginica":3})
```

```python
sns.pairplot(df,hue='species')
```

```python
#splitting data into training and testing part
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
X=df.drop(["species"],axis="columns")#explanatory variables
y=df.species
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
#80% for train and 20% for testing
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_train,y_train)#fitting the model
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[6,3,5,2]])
```
